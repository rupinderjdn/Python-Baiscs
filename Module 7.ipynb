### 14) HANDLING ‘Nan’ VALUES


#### 1) Creating nan values

```python
df.head()
```

```python
df=iris.copy()
```

```python
df.head()
```

```python
import numpy as np
df.iloc[2:4,1:3]=np.nan
df.head()
```

#### 2) Removing nan values

```python
df.dropna(inplace=True)
df.head()
```

#### 3) Filling nan values

```python
df.head()
```

```python
df['SepalWidthCm'].fillna(df['SepalWidthCm'].mean(),inplace=True)  
df['PetalLengthCm'].fillna(df['PetalLengthCm'].mean(),inplace=True)
df.head()
```

```python
a=df[df['Species']=='Setosa'] 
sepal_mean=a['SepalWidthCm'].mean() 
petal_mean=a['PetalLengthCm'].mean()
df['SepalWidthCm'].fillna(sepal_mean,inplace=True) 
df['PetalLengthCm'].fillna(petal_mean,inplace=True) 
df.head()
```
